```python
# notebooks/02_feature_engineering.ipynb

import pandas as pd
import sys
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Add src to path to import our modules
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

import config
from src.data_processing import load_and_clean_data, create_features

# --- 1. Load Clean Data ---
# Load cleaned data from the previous step
try:
    df_cleaned = load_and_clean_data(config.RAW_DATA_FILE)
    print("Cleaned data loaded successfully.")
except FileNotFoundError:
    print(f"Error: Make sure '{config.RAW_DATA_FILE}' exists and notebook 01 has been run.")
    raise

# --- 2. Create New Advanced Features ---
# The new create_features function doesn't need extra parameters
print("Creating advanced features...")
df_featured = create_features(df_cleaned)

# --- 3. Save Processed Data ---
# New, correct way
df_featured.to_csv(config.PROCESSED_DATA_FILE, index=True)
print(f"Processed data with features saved to {config.PROCESSED_DATA_FILE}")
print("\nFeatured DataFrame Head:")
print(df_featured.head())
print("\nFeatured DataFrame Tail:")
print(df_featured.tail())
print("\nFeatured DataFrame Info:")
df_featured.info()


# --- 4. Visualize Some of the New Features ---
print("\nVisualizing new features...")
fig, axes = plt.subplots(3, 1, figsize=(20, 15), sharex=True)

# Plot 1: HMA and VWAP vs. Close Price
df_featured[['Close', 'hma', 'vwap']].tail(500).plot(
    ax=axes[0],
    title='Price vs. Hull Moving Average (HMA) and VWAP'
)
axes[0].grid(True, linestyle='--', alpha=0.5)

# Plot 2: RSI and ADX (Momentum and Trend Strength)
df_featured[['rsi', 'adx']].tail(500).plot(
    ax=axes[1],
    title='RSI and ADX',
    secondary_y=['adx'] # Put ADX on a secondary axis if scales differ
)
axes[1].grid(True, linestyle='--', alpha=0.5)


# Plot 3: Normalized ATR and BB Width (Volatility Indicators)
df_featured[['atr_norm', 'bb_width_norm']].tail(500).plot(
    ax=axes[2],
    title='Normalized Volatility (ATR & Bollinger Band Width)'
)
axes[2].grid(True, linestyle='--', alpha=0.5)


plt.tight_layout()
plt.show()
```
